## Comparación de Modelos: Naïve Bayes vs Árbol de Decisión vs Random Forest

### **Regresión: Naïve Bayes vs Random Forest**

Para evaluar qué modelo es mejor para predecir los **precios de las viviendas**, se compararon los errores obtenidos en cada modelo:

| Modelo            | MAE ↓                 | MSE ↓                        | RMSE ↓                |
| ----------------- | --------------------- | ---------------------------- | --------------------- |
| **Naïve Bayes**   | 34,056.99             | 3,803,718,999.20             | 61,674.30             |
| **Random Forest** | **(Mejor)** 26,527.59 | **(Mejor)** 1,755,940,028.20 | **(Mejor)** 41,903.94 |

**Análisis de Resultados:**

- **Random Forest superó significativamente a Naïve Bayes en todas las métricas de error**.
- La diferencia en **MAE (34,056 vs 26,527)** indica que en promedio, Random Forest tiene errores menores en la predicción del precio de una casa.
- El **MSE y RMSE** confirman que Random Forest tiene mejor ajuste y menor dispersión en sus predicciones.

**Conclusión:**  
Para la tarea de **regresión**, **Random Forest es la mejor opción** debido a su menor error en la predicción del precio de las viviendas. Naïve Bayes no es el mejor algoritmo para predicciones continuas en este caso.

---

### **Clasificación: Naïve Bayes vs Árbol de Decisión**

Ahora, comparamos el rendimiento de los modelos para clasificar las viviendas en **Económica, Intermedia o Cara**.

| Modelo                | Accuracy ↑ | Precision ↑ | Recall ↑  | F1-Score ↑ |
| --------------------- | ---------- | ----------- | --------- | ---------- |
| **Naïve Bayes**       | 0.82       | 0.82        | 0.82      | 0.82       |
| **Árbol de Decisión** | 0.82       | 0.79-0.83   | 0.69-0.87 | 0.71-0.85  |

**Análisis de Resultados:**

- **Ambos modelos obtuvieron una precisión global de 82%**, lo que significa que aciertan en el mismo porcentaje de clasificaciones.
- Sin embargo, en categorías específicas, el **Árbol de Decisión logra un mejor recall en Económica (87%) y mejor F1-Score (85%) en comparación con Naïve Bayes**.
- Naïve Bayes tiende a clasificar de manera más homogénea, mientras que el Árbol de Decisión logra distinguir mejor algunos grupos, aunque puede sobreajustarse en otros.

**Conclusión:**  
Para clasificación, **ambos modelos tienen el mismo rendimiento global**, pero el **Árbol de Decisión tiene mejor desempeño en ciertas categorías**. En términos de estabilidad, **Naïve Bayes es más balanceado**, pero **el Árbol de Decisión ofrece más precisión en ciertas clases**.

### **Comparación de Tiempos de Procesamiento**

| Modelo                | Tiempo de Entrenamiento |
| --------------------- | ----------------------- |
| **Naïve Bayes**       | **Muy Rápido**          |
| **Árbol de Decisión** | Moderado                |
| **Random Forest**     | **Más lento**           |

**Análisis de Eficiencia:**

- **Naïve Bayes es el más rápido** debido a su simplicidad matemática. Es ideal para modelos que deben ser ejecutados repetidamente en tiempo real.
- **El Árbol de Decisión es intermedio**, ya que construye un árbol de decisiones basado en los datos.
- **Random Forest es el más lento**, porque entrena múltiples árboles y promedia sus resultados, pero también es el más preciso.

---

### **Conclusión Final**

- Para **regresión**, **Random Forest es el mejor modelo** en términos de precisión, aunque es más lento en procesamiento.
- Para **clasificación**, **Naïve Bayes y el Árbol de Decisión tienen precisión similar**, pero el Árbol de Decisión es más detallado en categorías específicas.
- **Si el tiempo de entrenamiento es un factor importante**, **Naïve Bayes es la mejor opción**, pero si la prioridad es precisión, **Random Forest es la mejor elección**.

**Elección recomendada:**

- **Si se prioriza velocidad:** Naïve Bayes.
- **Si se prioriza precisión:** Random Forest para regresión y Árbol de Decisión para clasificación.
