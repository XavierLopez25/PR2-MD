Se utilizaron dos modelos:

1. **MRL Univariado:** utiliza únicamente `OverallQual` para predecir `log(SalePrice)`.
2. **MRL Multivariable:** utiliza múltiples variables predictoras relevantes seleccionadas mediante análisis de correlación.

Ambos modelos fueron entrenados con el 70% de los datos (conjunto de entrenamiento) y posteriormente evaluados con el 30% restante (conjunto de prueba).

### Métricas utilizadas para evaluar la eficiencia

Para medir la eficiencia de cada modelo se utilizaron:

| Métrica               | Descripción                                                          |
| --------------------- | -------------------------------------------------------------------- |
| **R²**                | Proporción de la varianza explicada por el modelo.                   |
| **MSE**               | Error cuadrático medio, refleja la magnitud promedio de los errores. |
| **Máxima diferencia** | Máximo error absoluto observado (caso más difícil para el modelo).   |

Además de la gráfica de de dispersión residual y el histograma realizado para ambos modelos MRL.

### **Evaluación MRL Univariada (OverallQual como único predictor)**

#### ¿Qué tan bien lo hizo?

El modelo de regresión lineal univariada mostró un desempeño **moderado**. Logró explicar el **69%** de la variabilidad de `log(SalePrice)` (**R² = 0.69**), lo cual es aceptable considerando que solo utiliza una variable _(OverallQual)_. Sin embargo, este nivel de precisión es **limitado** para un mercado inmobiliario complejo, donde múltiples factores influyen en el precio de una vivienda.

El **MSE obtenido fue 0.05**, lo cual refleja que, en promedio, las predicciones tienen un error cuadrático relativamente alto. Además, la **máxima diferencia observada** entre una predicción y el valor real alcanzó **0.94 unidades logarítmicas**, indicando que el modelo puede fallar notablemente en propiedades atípicas o fuera de patrón.

### **Evaluación MRL Multivariable (modelo con 8 variables predictoras)**

#### ¿Qué tan bien lo hizo?

El modelo de regresión lineal multivariable mostró un desempeño **notablemente superior**. Logró explicar **84%** de la variabilidad de `log(SalePrice)` (**R² = 0.84**), lo cual es un salto significativo respecto al modelo univariado. Este alto R² muestra que el modelo es capaz de capturar mejor la **complejidad real del mercado inmobiliario**.

El **MSE fue 0.03**, indicando que en promedio las predicciones son bastante cercanas a los valores reales. La **máxima diferencia observada fue 0.92**, similar al modelo univariado, lo cual es lógico, ya que sigue habiendo **outliers** y propiedades excepcionales difíciles de modelar. Sin embargo, el multivariable lo hace **mejor en la mayoría de los casos**, reduciendo la magnitud general de los errores.

### Análisis Gráfico MRL Univariada

#### Predicción vs Reales para MRL Univariado



Esta gráfica muestra la relación entre OverallQual (calidad general de la construcción) y log(SalePrice) para el conjunto de prueba. Se observa una relación lineal positiva clara, lo cual valida que OverallQual es un predictor relevante del precio de las casas: a mayor calidad, mayor precio. Sin embargo, se evidencia una considerable dispersión vertical alrededor de la línea de regresión, lo cual indica que, aunque OverallQual es una variable importante, no es suficiente para explicar completamente el precio de las viviendas. Hay una gran variabilidad en los precios reales dentro de cada nivel de calidad, reflejando que hay factores adicionales (como el tamaño, la ubicación o las amenidades) que el modelo univariado no captura.

#### Dispersión Residual



Esta gráfica es clave para analizar si el modelo cumple con el supuesto de homocedasticidad, es decir, si el tamaño de los errores es constante a lo largo de todo el rango de valores de OverallQual. En este caso, se observa que la dispersión de los residuales es mayor en propiedades con alta calidad, lo que sugiere heterocedasticidad. Esto significa que el modelo es relativamente más preciso al predecir propiedades de baja o media calidad, pero pierde precisión al predecir propiedades de alta calidad, probablemente porque estas dependen de características adicionales que no están en el modelo, como acabados de lujo, ubicación premium, o características especiales. A pesar de esto, la ausencia de un patrón curvo indica que la relación lineal es razonablemente adecuada, aunque incompleta.

#### Histograma de Residuales



El histograma de residuales revela la distribución de los errores del modelo. En este caso, los residuales siguen una forma aproximadamente normal, lo cual es favorable. Sin embargo, se observa una leve asimetría hacia la izquierda (cola izquierda), lo que indica que el modelo tiende a subestimar el precio de algunas viviendas. Esta subestimación ocurre principalmente en propiedades de alta calidad o características premium, que suelen tener precios fuera del patrón regular del mercado. Esta gráfica confirma que, si bien el modelo captura la tendencia general, no es suficiente para capturar toda la complejidad del mercado inmobiliario.

#### MRL Multivariable

Al incorporar variables adicionales (`GrLivArea`, `GarageCars`, `YearBuilt`, entre otras), el modelo mejora significativamente su capacidad predictiva, alcanzando un **R² de 0.84** y reduciendo el **MSE** a 0.03. Esto demuestra que **el precio de las casas es un fenómeno multifactorial**, y considerar solo una variable no es suficiente.

### Conclusión

Con base en el análisis cuantitativo (métricas) y cualitativo (gráficos), se puede concluir lo siguiente:

- El **modelo multivariable es claramente superior** al univariado.
- Explica una **mayor proporción de la variabilidad** (R² = 0.84 vs 0.69).
- Produce **menores errores en promedio** (MSE = 0.03 vs 0.05).
- Presenta una **mejor distribución de residuos**, con menor evidencia de patrones sistemáticos.

El modelo univariado, aunque sencillo y fácil de interpretar, es **demasiado simplista** para un mercado complejo como el inmobiliario. Solo capturar la calidad general _(OverallQual)_ **no es suficiente**, pues el precio real está influenciado por múltiples factores adicionales como el tamaño, la antigüedad y las amenidades.

En conclusión, **el mejor modelo para predecir el precio de las casas es el MRL multivariable**, ya que logra el mejor balance entre capacidad predictiva y robustez, aprovechando la sinergia entre varias variables clave.

### Reflexión Final

Este ejercicio mostró que, aunque `OverallQual` es un predictor fuerte, el precio de las viviendas es un fenómeno **multifactorial**. Ignorar variables como el tamaño, el año de construcción o el garaje, conduce a un modelo incompleto. El multivariable es más robusto porque considera **múltiples factores a la vez**, lo que refleja mejor la realidad del mercado inmobiliario.
